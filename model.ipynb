```python
##data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

## Data preprocessing

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
```

```python
# removing stop words and stemming the data
ps = PorterStemmer()
corpus = []

# Assuming 'data' is your DataFrame containing the 'v2' column
for i in range(0, len(data)):
    review = re.sub('[^a-zA-Z]', ' ', data['v2'][i])  # Fixed the regex pattern
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]  # Fixed the condition
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create Bag of Words

```python
y=pd.get_dummies(data['v1'])
y=y.iloc[:,0].values
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train),len(y_train)
```

```python
## independent features
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

## Classification using MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## Creating The TF_IDF Model

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
import joblib

# Assume `model` is your trained model
model = tv 

# Save the model as a .pkl file
joblib.dump(model, 'tfidf_vectorizer.pkl')
```
